```python
import pandas as pd
import pandera as pa
```

```python
df = pd.read_csv("ocorrencia.csv", parse_dates=['ocorrencia_dia'], dayfirst=True, infer_datetime_format=True)
df.head(5)
```

```python
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
```

```python
df.head()
```

```python
df.replace(['**', '###!', '####', '****', '*****', 'NULL'], pd.NA, inplace=True)
```

```python
## informa se o valor do campo é NA (true/false)
# df.isna()

# informa a soma da quantidade de valores NA no dataframe por coluna
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
# altera todos os valores NA ou NULL para o valor informado
df.fillna(10, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# substitui todos os valores 10 por NA
df.replace([10], pd.NA, inplace=True)
```

```python
df.isnull().sum()
```

```python
# substitui todos os valores NA por 10 na coluna especificada
df.fillna(value={'total_recomendacoes':10}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['total_recomendacoes_bkp'] = df.total_recomendacoes 
```

```python
df.head()
```

```python
# remove a coluna especificada
df.drop(['total_recomendacoes_bkp'], axis=1, inplace=True)
```

```python
# remove todas as linhas que possuem o valor NA
df.dropna()
```
